```python
from tika import parser
import numpy as np
import string
```

```python
allowed=string.ascii_letters + " "
```

```python
rawText = parser.from_file("21SEPT-USSF-GUARDIAN-IDEAL.pdf")
```

```python
text = " ".join(rawText['content'].split())
sentences = text.split(".")
```

```python
sentences = [s.strip() for s in sentences]
```

```python
m = np.mean([len(s) for s in sentences])
sd = np.std([len(s) for s in sentences])
```

```python
sentences = [s for s in sentences if len(s) > (m - sd)]
```

```python
for idx in range(len(sentences)):
    sentences[idx] = "".join([s for s in sentences[idx] if s in allowed])
    sentences[idx] = " ".join([s for s in sentences[idx].split() if len(s) > 3])
    
```

```python
with open("ideal.txt", 'w') as file_out:
    file_out.writelines("\n".join(sentences))
```
